```python
import pandas as pd
from PIL import Image, ImageDraw, ImageFilter, ImageFont
```

```python
file = r"\Stadt.csv"
data = pd.read_csv(file,sep=';')
```

```python
values = []
for i in range (len(data)):
    values.append([])
    for j in range(1,5):
        values[i].append(data.iloc[i][j] == "-")
    
```

```python
def f_isim(x):
    im2_name = "Ön/Photos/"+str(x)+".png"
    im1 = Image.open("Arka/İsim/isim.png").convert("RGBA")
    im2 = Image.open(im2_name).convert("RGBA")

    msg1 = data.iloc[x][1]
    msg2 = data.iloc[x][2]
    msg3 = data.iloc[x][3]
    
    layer2 = im2.resize((80,80))
    layer3 = im2.resize((80,80))
    
    width, height = im1.size
    layer2.paste(layer3, (20,10), mask=layer3)
    w, h = layer2.size
    
    im1.paste(layer2, (10,380), mask=layer2)

    draw = ImageDraw.Draw(im1)
    font = ImageFont.truetype("AftaSansThin-Regular.otf", 13)
    w,h = font.getsize(msg1)
    draw.text((130, 95-h),msg1,(0,0,0),font=font)
    
    draw = ImageDraw.Draw(im1)
    font = ImageFont.truetype("AftaSansThin-Regular.otf", 13)
    w,h = font.getsize(msg2)
    draw.text((130, 265-h),msg2,(0,0,0),font=font)
    
    draw = ImageDraw.Draw(im1)
    font = ImageFont.truetype("AftaSansThin-Regular.otf", 13)
    w,h = font.getsize(msg3)
    draw.text((130, 425-h),msg3,(0,0,0),font=font)

    name = "Arka/photo_"+str(i)+".png"
    im1.save(name)
```

```python
def f_fiil(x):
    im2_name = "Ön/Photos/"+str(i)+".png"
    im1 = Image.open("Arka/Fiil/fiil.png").convert("RGBA")
    im2 = Image.open(im2_name).convert("RGBA")

    msg1 = data.iloc[x][1]
    msg2 = data.iloc[x][2]
    msg3 = data.iloc[x][4]   
   
    width, height = im1.size

    draw = ImageDraw.Draw(im1)
    font = ImageFont.truetype("AftaSansThin-Regular.otf", 13)
    w,h = font.getsize(msg1)
    draw.text((130, 70),msg1,(0,0,0),font=font)
    
    draw = ImageDraw.Draw(im1)
    font = ImageFont.truetype("AftaSansThin-Regular.otf", 13)
    w,h = font.getsize(msg2)
    draw.text((130, (height/2)-h+10),msg2,(0,0,0),font=font)
    
    msg3_a = msg3.split()
    leng = len(msg3_a)
    for a in range (leng):
        if (a<leng/2):
            draw = ImageDraw.Draw(im1)
            font = ImageFont.truetype("AftaSansThin-Regular.otf", 20)
            w,h = font.getsize(msg3_a[a])
            draw.text((30+200*(a/2), 380),msg3_a[a],(255,102,102),font=font)
            #draw.text((80*(a+1)-50, 380),msg3_a[a],(255,102,102),font=font)
            
        if (a>=leng/2):
            draw = ImageDraw.Draw(im1)
            font = ImageFont.truetype("AftaSansThin-Regular.otf", 15)
            w,h = font.getsize(msg3_a[a])
            draw.text((30+200*(a-6)/2,410+h),msg3_a[a],(0,0,0),font=font)
            #draw.text((80*(a-5)-50, 410+h),msg3_a[a],(0,0,0),font=font)
    name = "Arka/photo_"+str(i)+".png"
    im1.save(name)
    #name = "Arka/photo_"+str(i)+".jpg"
    #im1.save(name)
```

```python
def f_ja(x):
    im2_name = "Ön/Photos/"+str(i)+".png"
    im1 = Image.open("Arka/ja/ja.png").convert("RGBA")
    
    msg = data.iloc[x][2]
    width, height = im1.size

    draw = ImageDraw.Draw(im1)
    font = ImageFont.truetype("AftaSansThin-Regular.otf", 40)
    w,h = font.getsize(msg)
    draw.text((int((width-w)/2),int((height-h)/2)-10),msg,(0,0,0),font=font)

    name = "Arka/photo_"+str(i)+".png"
    im1.save(name)
```

```python
def f_soru(x):
    im2_name = "Ön/Photos/"+str(i)+".png"
    im1 = Image.open("Arka/soru/soru.png").convert("RGBA")
    
    msg1 = data.iloc[x][1]
    msg2 = data.iloc[x][2]
    
    draw = ImageDraw.Draw(im1)
    font = ImageFont.truetype("AftaSansThin-Regular.otf", 25)
    w,h = font.getsize(msg1)
    draw.text((150,150-h),msg1,(0,0,0),font=font)    

    draw = ImageDraw.Draw(im1)
    font = ImageFont.truetype("AftaSansThin-Regular.otf", 25)
    w,h = font.getsize(msg2)
    draw.text((150,390-h),msg2,(0,0,0),font=font)
    name = "Arka/photo_"+str(i)+".png"
    im1.save(name)
```

```python
def f_on(x):
    im2_name = "Ön/Photos/"+str(x)+".png"
    im1 = Image.open("Ön/Photos/on.png").convert("RGBA")
    im2 = Image.open(im2_name).convert("RGBA")

    msg = data.iloc[x][0]

    #layer1 = im1.copy()
    layer2 = im2.resize((210, 210))

    width, height = im1.size
    w, h = layer2.size
    
    im1.paste(layer2, (int((width-w)/2),int((height-h)/2)+30), mask=layer2)

    draw = ImageDraw.Draw(im1)
    font = ImageFont.truetype("AftaSansThin-Regular.otf", 40)
    w,h = font.getsize(msg)
    draw.text(((width-w)/2, 40),msg,(255,255,255),font=font)

    name = "Ön/photo_"+str(x)+".png"
    
    if values[x] != [1,1,1,1]:
        im3 = Image.open("Ön/Photos/click.png").convert("RGBA")
        im3 = im3.resize((80,80))
        w, h = im3.size
        im1.paste(im3, (int((width-w)/2),450), mask=im3)
    else:
        pass
    
    im1.save(name)
```

```python
def f_cogul(x):
    im2_name = "Ön/Photos/"+str(x)+".png"
    im1 = Image.open("Arka/cogul/cogul.png").convert("RGBA")
    im2 = Image.open(im2_name).convert("RGBA")
    
    msg = data.iloc[x][3]
    width, height = im1.size
    
    layer2 = im2.resize((150,150))
    layer3 = im2.resize((150,150))
    
    width, height = im1.size
    layer2.paste(layer3, (40,20), mask=layer3)
    w, h = layer2.size
    
    im1.paste(layer2, (int((width-w)/2),int(-160+(height-h)/2)), mask=layer2)

    draw = ImageDraw.Draw(im1)
    font = ImageFont.truetype("AftaSansThin-Regular.otf", 35)
    w,h = font.getsize(msg)
    draw.text((int((width-w)/2),int((height-h)/2)),msg,(0,0,0),font=font)

    name = "Arka/photo_"+str(x)+".png"
    
    im1.save(name)
    
```

```python
for i in range (len(data)):
    f_on(i)
    if values[i] == [0,0,0,1]: #İsim        
        f_isim(i)
        continue
        
    if values[i] == [0,0,1,0]: #fiil        
        f_fiil(i)
        continue
        
    if values[i] == [1,0,1,1]: #ja        
        f_ja(i)
        continue
        
    if values[i] == [1,1,0,1]: #cogul        
        f_cogul(i)
        continue
        
    if values[i] == [0,0,1,1]: #soru        
        f_soru(i)
        continue
```
