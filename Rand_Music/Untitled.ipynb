```python
from random import randrange
import cv2
import numpy as np
from pydub import AudioSegment
from pydub.playback import *
import time
import threading
def random_notes(x):
    starter = AudioSegment.silent(duration=10)
    
    links = [r"\Notalar\1\Do.png",
            r"\Notalar\1\Re.png",
            r"\Notalar\1\Mi.png",
            r"\Notalar\1\Fa.png",
            r"\Notalar\2\Sol.png",
            r"\Notalar\2\La.png",
            r"\Notalar\2\Si.png",
            r"\Notalar\2\Do2.png",
            r"\Notalar\3\Re2.png",
            r"\Notalar\3\Mi2.png",
            r"\Notalar\3\Fa2.png",
            r"\Notalar\3\Sol2.png",
            r"\Notalar\4\La2.png",
            r"\Notalar\4\Si2.png",
            r"\Notalar\4\Do3.png",
            r"\Notalar\4\Re3.png"]
    
    Soundlink = [r"\Notalar\1\Do1.wav",
            r"\Notalar\1\Re1.wav",
            r"\Notalar\1\Mi1.wav",
            r"\Notalar\1\Fa1.wav",
            r"\Notalar\2\Sol1.wav",
            r"\Notalar\2\La1.wav",
            r"\Notalar\2\Si1.wav",
            r"\Notalar\2\Do2.wav",
            r"\Notalar\3\Re2.wav",
            r"\Notalar\3\Mi2.wav",
            r"\Notalar\3\Fa2.wav",
            r"\Notalar\3\Sol2.wav",
            r"\Notalar\4\La2.wav",
            r"\Notalar\4\Si2.wav",
            r"\Notalar\4\Do3.wav",
            r"\Notalar\4\Re3.wav"]
    
    notes = ["Do","Re","Mi","Fa","Sol","La","Si","Do","Re","Mi","Fa","Sol","La","Si","Do","Re"]
    files = []
    rn = []
    r_notes = []
    conc_pre = []
    for i in range (x):
        rn.append(randrange(16))

    for i in range(x):
        files.append(cv2.imread(links[rn[i]]))
        r_notes.append(notes[rn[i]])
        starter = starter + AudioSegment.from_wav(Soundlink[rn[i]])
        
    for i in range(x):                
        conc_pre.append(files[i])
        conc_pre[i].shape
    
    numpy_horizontal_concat = np.concatenate((conc_pre), axis=1)
    
    def finalimage():
        cv2.imshow(f"{r_notes}", numpy_horizontal_concat)
        cv2.waitKey()
        time.sleep(0.5)
    thread = threading.Thread(target=finalimage)
    thread.start()
    play(starter)
    
    
    
    


def main():
    Flag = 0
    input1 = int(input("Kaç nota istersin?"))
    if input1 >= 26:
        print("26 dan küçük girmelisin.")
        main()
    random_notes(input1)
    main()
main()
```
